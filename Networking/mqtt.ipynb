```python
import paho.mqtt.client as mqtt
```

```python
topic = 'test/esp32'
broker = '192.168.1.101'
port = 1883

username = 'testUser'
password = 'Abc_072179'
```

```python
# Define callback functions
def on_connect(client, userdata, flags, rc):
    print('Connected with result code: ' + str(rc))
#     client.subscribe(topic)

def on_publish(client, userdata, mid):
    print('Published message')

def on_message(client, userdata, message):
    print('Received message: ' + message.payload.decode())
```

## Publisher

```python
publisher = mqtt.Client(client_id='publisher-client')
# Set callbacks
publisher.on_connect = on_connect
publisher.on_publish = on_publish
# publisher.on_message = on_message
# Connect to broker
publisher.username_pw_set(username, password)
```

## Publish message with QoS=2

```python
# Publish test
publisher.connect(broker, port)
# publisher.loop_start()
# publisher.loop_stop()

data_chunk = b'\x0ff'
for i in range(5000):
    publisher.publish(topic, data_chunk, qos=2)
#     publisher.loop_start()
#     publisher.loop_stop()
```
